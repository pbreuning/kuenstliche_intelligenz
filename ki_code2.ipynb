## Datengenerierung

```python
x = np.array([1.0, 2.0])
y = np.array([300.0, 500.0])

m = x.shape[0]

# Alternativ: m = len(x)


print(f"x: { x }, y: { y }, m={ m }")
```

## Werte des Modells berechnen

```python
def compute_model_output(x, w, b):
# x: Array,   w,b: Skalare
# Gewünschte Ausgabe: Array mit Werten des Modells

    m = x.shape[0]   
    # Ist Länge des Arrays

    f_wb = np.zeros(m)
    # Array mit m Nullen

    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb

print(compute_model_output(x, 100, 150))
```

```python
tmp_f_wb = compute_model_output(x, 100, 150)

# Modell-Vorhersage plotten:
plt.plot(x, tmp_f_wb, c="b", label="Vorhersage")

# Die Daten plotten:
plt.scatter(x, y, marker="x", c="r")

# Titel setzen
plt.title("Hauspreise")

# x-Achsen-Label:
plt.xlabel("Größe (in 100 qm)")

# y-Achsen-Label:
plt.ylabel("Preis (in 1000 Euro)")

# Label anzeigen:
plt.legend()

# Plot anzeigen:
plt.show()
```

## Wert der Kostenfunktion berechnen

```python
def compute_cost(x, y, w, b):

    # Anzahl der Trainingsbeispiele
    m = x.shape[0]

    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
    total_cost = (1 / (2 * m)) * cost_sum

    return total_cost

print(f"Kosten: {compute_cost(x, y, 100, 150)}")
```

## Gradient berechnen

```python
def compute_gradient(x, y, w, b):

    # Anzahl der Trainingsbeispiele
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db

print(f"Gradient: {compute_gradient(x, y, 100, 150)}")
```
